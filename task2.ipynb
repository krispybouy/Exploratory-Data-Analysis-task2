```python
# Imports
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
csv_path = "cleaned_titanic.csv"
if not os.path.exists(csv_path):
    raise FileNotFoundError(f"Cannot find file: {csv_path}")
df = pd.read_csv(csv_path)

csv_path2="cleaned_titanic_unscaled.csv"
if not os.path.exists(csv_path2):
    raise FileNotFoundError(f"Cannot find file: {csv_path2}")
df2 = pd.read_csv(csv_path2)
```

```python
#df.describe()
df2.describe()
```

```python
df2.boxplot(column='Age', by='Survived')
plt.title('Boxplot of Age by Survival Status')
df2.hist(column='Age', bins=20)
plt.title('Histogram of Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histograms for numeric columns
df2.hist(figsize=(12, 8), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Numeric Features")
plt.show()

```

```python
# Correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(df2.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

# Pairplot for visual feature relationships
sns.pairplot(df2, diag_kind='kde')
plt.suptitle("Pairplot of Numeric Features", y=1.02)
plt.show()
```

```python
# Example: Survival rate vs class
sns.barplot(data=df, x='Pclass', y='Survived')
plt.title("Survival Rate by Passenger Class")
plt.show()

# Example: Survival by gender
sns.barplot(data=df, x='Sex_male', y='Survived')
plt.title("Survival Rate by Gender")
plt.show()
```

Key Observations 

- passenders in higher classes had higher survival rates
- females had significantly better survival odds than males
- age distribution is right-skewed which means most passengers were between 20-40
- fare has noticebale outliers
- strong correlation between passenger class and fare
